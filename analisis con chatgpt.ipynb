```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('ventas.csv',encoding= 'utf-8')
data
```

```python
# Limpieza y creación de métricas
data['Fecha'] = pd.to_datetime(data['Fecha de venta'])
data['Mes'] = data['Fecha'].dt.month
data['Año'] = data['Fecha'].dt.year
data['Ingresos'] = data['Cantidad vendida'] * data['Precio unitario']
```

```python
data
```

```python
productos = data.groupby('Producto')['Ingresos'].sum().sort_values(ascending=False)
```

```python
productos
```

```python
tendencia = data.groupby(['Año', 'Mes'])['Ingresos'].sum()
tendencia
```

```python
plt.figure(figsize=(10,5))
productos.head(10).plot(kind= 'bar', title='Productos más vendidos')
plt.show()
```

```python
plt.figure(figsize=(10,5))
tendencia.plot(kind='line', title='Tendencia de Ventas Mensuales')
plt.show()
```

```python
file_path = "archivo_venta_limpiado.csv"
data.to_csv(file_path, index= False, encoding='utf-8-sig')
file_path
```
